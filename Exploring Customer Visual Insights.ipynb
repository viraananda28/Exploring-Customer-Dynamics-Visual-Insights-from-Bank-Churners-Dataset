1 dan 2.  Visualisasi dan insight untuk kolom attrition_flag, customer_age,
gender, education_level, dan income_category

```python
# Visualisasi untuk kolom attrition_flag
# Pie Chart
fig, ax = plt.subplots(1, 2, figsize=(14, 7))
attrition_counts = df['Attrition_Flag'].value_counts()
ax[0].pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90)
ax[0].set_title('Attrition Flag Distribution (Pie Chart)')

# Donut Chart
wedges, texts, autotexts = ax[1].pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
ax[1].set_title('Attrition Flag Distribution (Donut Chart)')

plt.show()
```

Attrition flag didominasi oleh existing customer dengan persentase yang sangat besar yaitu 83,9%. Artinya pelanggan puas dengan layanan yang diberikan. Sementara itu 16.1% merupakan attrited customer, sehingga perusahaan perlu ditindaklanjuti terkait faktor yang menyebabkan pelanggan berhenti berlangganan.

```python
# Visualisasi untuk kolom customer_age
# Vertical Bar Chart
plt.figure(figsize=(10, 6))
sns.histplot(df['Customer_Age'], kde=False, bins=30)
plt.title('Customer Age Distribution (Vertical Bar Chart)')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Pelanggan yang paling banyak menggunakan layanan adalah pelanggan yang berada di rentang umur setengah baya, yaitu 40-55 tahun. Kelompok usia ini mungkin memiliki kebutuhan keuangan yang lebih kompleks dan stabil, sehingga mereka cenderung lebih setia dan menggunakan berbagai produk dan layanan yang ditawarkan oleh bank. Perusahaan bisa fokus pada segmen umur ini dengan menawarkan produk dan layanan yang khusus dirancang untuk memenuhi kebutuhan mereka.

```python
# Visualisasi untuk kolom gender
# Horizontal Bar Chart
plt.figure(figsize=(10, 6))
gender_counts = df['Gender'].value_counts()
sns.barplot(x=gender_counts.values, y=gender_counts.index, palette='plasma')
plt.title('Gender Distribution (Horizontal Bar Chart)')
plt.xlabel('Count')
plt.ylabel('Gender')
plt.show()
```

Pelanggan yang paling banyak menggunakan layanan adalah female yaitu sekitar 5700 orang, sedangkan jumlah pelanggan male tidak memiliki perbedaan yang signifikan, yaitu sekitar 4800 orang. Perusahaan dapat mempertimbangkan untuk mengembangkan produk dan layanan yang lebih sesuai atau menarik bagi female dan male, serta mempertimbangkan strategi pemasaran yang ditargetkan untuk menarik lebih banyak pelanggan dari kedua gender tersebut.

```python
# Visualisasi untuk kolom education_level
# Vertical Bar Chart
plt.figure(figsize=(10, 6))
edu_counts = df['Education_Level'].value_counts()
sns.barplot(x=edu_counts.index, y=edu_counts.values, palette='plasma')
plt.title('Education Level Distribution (Vertical Bar Chart)')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

layanan bank menarik bagi berbagai tingkat pendidikan, tetapi sebagian besar pengguna adalah lulusan sarjana. Perusahaan dapat menggunakan informasi ini untuk mengembangkan program dan layanan yang lebih baik sesuai dengan tingkat pendidikan pelanggan. Misalnya, layanan yang lebih kompleks dan bervariasi mungkin lebih sesuai untuk mereka yang berpendidikan tinggi, sementara pendekatan yang lebih sederhana bisa lebih menarik bagi mereka dengan tingkat pendidikan yang lebih rendah.

```python
# Visualisasi untuk kolom income_category
# Pie Chart
plt.figure(figsize=(10, 6))
income_counts = df['Income_Category'].value_counts()
plt.pie(income_counts, labels=income_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('plasma'))
plt.title('Income Category Distribution (Pie Chart)')
plt.show()
```

Layanan bank sangat menarik bagi sebagian besar pelanggan yang berada dalam kategori pendapatan rendah hingga menengah. Karena pelanggan yang berada dalam kategori pendapatan di bawah 40K paling banyak menggunakan layanan bank, maka perusahaan dapat mengembangkan produk yang lebih terjangkau atau menawarkan program bantuan keuangan. Sedangkan untuk pelanggan yang memiliki pendapatan menengah hingga 120K+, perusahaan  dapat mengembangkan produk premium atau layanan khusus yang dapat menarik lebih banyak pelanggan dalam segmen ini.

3 dan 6. Visualisasi menggunakan countplot dan insight

```python
# Countplot memadukan kolom Gender dengan Attrition Flag
plt.figure(figsize=(14, 7))
sns.countplot(data=df, x='Gender', hue='Attrition_Flag', palette='coolwarm')
plt.title('Gender vs Attrition Flag Countplot')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Pelanggan female merupakan pelanggan tertinggi yang masih menggunakan layanan bank tetapi juga merupakan pelanggan tertinggi yang sudah tidak berlangganan. Perusahaan dapat fokus kepada pelanggan female untuk terus meningkatkan jumlah pelanggan dengan memberikan layanan tambahan yang menarik bagi female. Hal ini sekiranya dapat membuat pelanggan female tidak jadi menghentikan penggunaan layanan bank.

4 dan 6. Visualisasi menggunakan scatterplot dan insight

```python
#Scatterplot memadukan kolom customer age dengan credit limit
plt.figure(figsize=(14, 7))
sns.scatterplot(data=df, x='Customer_Age', y='Credit_Limit', hue='Attrition_Flag', palette='coolwarm')
plt.title('Customer Age vs Credit Limit Scatterplot')
plt.xlabel('Customer Age')
plt.ylabel('Credit Limit')
plt.show()
```

Informasi: Dari scatterplot terlihat bahwa baik pelanggan yang masih aktif (existing customers) maupun yang sudah berhenti (attrited customers) tersebar di seluruh rentang umur dan limit kredit. Pelanggan yang masih aktif tersebar lebih merata di berbagai tingkat limit kredit, dengan banyak dari mereka memiliki limit kredit yang lebih tinggi. Pelanggan yang berhenti menggunakan layanan cenderung memiliki limit kredit yang lebih rendah, meskipun ada juga beberapa yang memiliki limit kredit yang tinggi. Kelompok umur menengah (40-55 tahun) tampaknya memiliki limit kredit yang bervariasi. Banyak dari mereka yang masih aktif sebagai pelanggan menunjukkan bahwa mereka mungkin merupakan target pasar utama bank.

Insight: Bank dapat mempertimbangkan untuk menawarkan produk yang lebih sesuai dengan berbagai kelompok umur. Misalnya, produk yang lebih menarik untuk pelanggan yang lebih muda dengan limit kredit lebih rendah, atau layanan premium untuk pelanggan yang lebih tua dengan limit kredit lebih tinggi.

5 dan 6. Visualisasi menggunakan heatmap dan insight

```python
# Heatmap memadukan lebih dari satu kolom
correlation_matrix = df[['Customer_Age', 'Credit_Limit', 'Total_Revolving_Bal', 'Avg_Open_To_Buy', 'Total_Trans_Amt', 'Total_Trans_Ct']].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

Informasi: Total transaksi dan jumlah transaksi memiliki korelasi positif yang sangat tinggi (mendekati 1). Ini menunjukkan bahwa semakin banyak jumlah transaksi yang dilakukan pelanggan, semakin tinggi jumlah total transaksi mereka. Hal ini logis karena jumlah transaksi yang lebih banyak akan meningkatkan total transaksi.
Saldo berputar dan rata-rata terbuka untuk dibeli memiliki korelasi negatif yang signifikan. Ini menunjukkan bahwa ketika saldo berputar meningkat, rata-rata jumlah yang tersedia untuk pembelian menurun. Hal ini masuk akal karena saldo berputar yang lebih tinggi berarti lebih banyak kredit yang digunakan, sehingga mengurangi jumlah kredit yang tersedia untuk pembelian.
Kemudian umur pelanggan tidak menunjukkan korelasi yang sangat kuat dengan kolom lain dalam heatmap ini. Ini menunjukkan bahwa umur pelanggan tidak memiliki hubungan yang kuat dengan batas kredit, saldo berputar, jumlah rata-rata yang terbuka untuk dibeli, total transaksi, atau jumlah transaksi.

Insight: Membantu dalam memahami hubungan antara berbagai variabel yang mempengaruhi perilaku pelanggan dalam menggunakan layanan bank. Dengan pemahaman ini, bank dapat mengembangkan strategi yang lebih efektif untuk mengelola kredit, meningkatkan penggunaan layanan, dan mengidentifikasi segmen pelanggan yang membutuhkan perhatian khusus. Misalnya, memahami bahwa pelanggan dengan jumlah transaksi yang tinggi cenderung memiliki total transaksi yang tinggi dapat membantu dalam merancang program loyalitas yang lebih tepat sasaran.
